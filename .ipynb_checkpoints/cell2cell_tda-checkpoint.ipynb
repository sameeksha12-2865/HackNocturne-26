## ⚙️ Cell 1 — Environment Check & Imports

````python
# ── Fix: Anaconda hides user site-packages. Add it BEFORE any imports. ────────
import sys, site, importlib
user_site = site.getusersitepackages()
if user_site not in sys.path:
    sys.path.insert(0, user_site)
print(f'Python: {sys.executable}')
print(f'User site-packages: {user_site}')

# ── Check for missing packages and install if needed ──────────────────────────
REQUIRED = {'kmapper': 'kmapper', 'umap': 'umap-learn', 'plotly': 'plotly',
            'networkx': 'networkx', 'sklearn': 'scikit-learn'}
missing_install = [pip for mod, pip in REQUIRED.items()
                   if importlib.util.find_spec(mod) is None]
if missing_install:
    import subprocess
    print(f'Installing: {missing_install}')
    subprocess.check_call([sys.executable, '-m', 'pip', 'install'] + missing_install + ['-q'])
    print('Done — restart kernel now (Kernel > Restart & Run All)')
else:
    print('✅ All packages present')

# ── Imports ───────────────────────────────────────────────────────────────────
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

import kmapper as km
from kmapper import Cover

from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.impute import SimpleImputer

try:
    from umap import UMAP
    HAS_UMAP = True
    print('✅ UMAP available — will use as primary lens')
except ImportError:
    HAS_UMAP = False
    print('⚠️  UMAP not found — falling back to PCA lens')

import networkx as nx
import plotly.graph_objects as go

np.random.seed(42)
print('✅ All imports OK')
````

---
## 📂 Cell 2 — Load Data

````python
# ── SET YOUR PATH HERE ────────────────────────────────────────────────────────
DATA_PATH = 'cell2celltrain.csv'   # change if your filename differs
# ─────────────────────────────────────────────────────────────────────────────

raw = pd.read_csv(DATA_PATH)
raw.columns = raw.columns.str.strip().str.lower().str.replace(' ', '_')

# Detect churn column
CHURN_COL = [c for c in raw.columns if 'churn' in c][0]
if raw[CHURN_COL].dtype == object:
    raw[CHURN_COL] = raw[CHURN_COL].str.strip().str.upper().map({'YES': 1, 'NO': 0})

print(f'Loaded:      {raw.shape[0]:,} rows x {raw.shape[1]} columns')
print(f'Churn col:   {CHURN_COL}')
print(f'Churn rate:  {raw[CHURN_COL].mean():.1%}')
raw.head(3)
````

---
## 🧹 Cell 3 — Preprocessing & Feature Engineering

Cell2Cell has **mixed dtypes** that must be handled before imputation:
- `object` Yes/No columns → encode to 1/0
- `homeownership`: Known/Unknown → 1/0  
- `creditrating`: ordinal string → integer rank
- `handsetprice`: stored as string numbers → coerce to float
- High-cardinality strings (`servicearea`, `prizmcode`, `occupation`, `maritalstatus`) → drop

After encoding, `FEATURE_COLS` is rebuilt from **actual dtypes** — so SimpleImputer shape mismatch is impossible.

````python
df = raw.copy()

# ── 1. handsetprice: object containing numeric strings → float ────────────────
if 'handsetprice' in df.columns:
    df['handsetprice'] = pd.to_numeric(df['handsetprice'], errors='coerce')

# ── 2. Yes/No object columns → 1/0 ───────────────────────────────────────────
for col in df.select_dtypes('object').columns:
    vals = df[col].dropna().str.strip().str.upper().unique()
    if set(vals).issubset({'YES', 'NO'}):
        df[col] = df[col].str.strip().str.upper().map({'YES': 1, 'NO': 0})
        print(f'  Encoded Yes/No: {col}')

# ── 3. homeownership: Known / Unknown → 1/0 ──────────────────────────────────
if 'homeownership' in df.columns and df['homeownership'].dtype == object:
    df['homeownership'] = df['homeownership'].str.strip().str.upper() \
                           .map({'KNOWN': 1, 'UNKNOWN': 0})
    print('  Encoded homeownership: Known=1 / Unknown=0')

# ── 4. creditrating: ordinal string → int rank ────────────────────────────────
if 'creditrating' in df.columns and df['creditrating'].dtype == object:
    credit_map = {
        '1-highest': 7, '2-high': 6, '3-good': 5, '4-medium': 4,
        '5-low': 3, '6-very low': 2, '7-lowest': 1
    }
    df['creditrating'] = df['creditrating'].str.strip().str.lower().map(credit_map)
    print('  Encoded creditrating: ordinal 1-7')

# ── 5. Drop remaining object columns (high-cardinality, not encodable) ────────
drop_obj = [c for c in df.select_dtypes('object').columns
            if c != CHURN_COL]
if drop_obj:
    print(f'  Dropping object columns: {drop_obj}')
    df = df.drop(columns=drop_obj)

# ── 6. Build FEATURE_COLS from actual numeric dtypes ─────────────────────────
FEATURE_COLS = [c for c in df.columns
                if c not in ['customerid', CHURN_COL]
                and pd.api.types.is_numeric_dtype(df[c])]

y = df[CHURN_COL].astype(int)
X_raw = df[FEATURE_COLS].copy()

print(f'\nFeature matrix:  {X_raw.shape[0]:,} rows x {X_raw.shape[1]} features')
print(f'All numeric:     {(X_raw.dtypes != object).all()}')
print(f'Missing values:  {X_raw.isnull().sum().sum():,}')
print(f'Churn rate:      {y.mean():.1%}')
````

---
## 🔧 Cell 4 — Impute & Scale

````python
# Impute — now safe because X_raw is guaranteed all-numeric
imputer = SimpleImputer(strategy='median')
X_imputed = pd.DataFrame(
    imputer.fit_transform(X_raw),
    columns=FEATURE_COLS,
    index=X_raw.index
)
print(f'Missing after imputation: {X_imputed.isnull().sum().sum()}')

# Scale
scaler = StandardScaler()
X_scaled = pd.DataFrame(
    scaler.fit_transform(X_imputed),
    columns=FEATURE_COLS,
    index=X_raw.index
)
print(f'✅ Feature matrix ready: {X_scaled.shape[0]:,} x {X_scaled.shape[1]}')
````

---
## 📊 Cell 5 — Why Global Models Fail (The Heterogeneity Problem)

````python
X_tr, X_te, y_tr, y_te = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y)

gbm_global = GradientBoostingClassifier(
    n_estimators=150, max_depth=4, subsample=0.8, random_state=42)
gbm_global.fit(X_tr, y_tr)
global_auc = roc_auc_score(y_te, gbm_global.predict_proba(X_te)[:, 1])
print(f'Global GBM AUC: {global_auc:.4f}')

# Churn probability for each row — used as Mapper lens
churn_proba = gbm_global.predict_proba(X_scaled)[:, 1]

# Top 20 feature importances
fi = pd.Series(gbm_global.feature_importances_, index=FEATURE_COLS).sort_values(ascending=False)

fig, ax = plt.subplots(figsize=(10, 7))
top20 = fi.head(20).sort_values()
colors = ['#e74c3c' if i >= 15 else '#3498db' for i in range(len(top20))]
top20.plot(kind='barh', ax=ax, color=colors)
ax.set_title(
    f'Global GBM Feature Importance  (AUC = {global_auc:.3f})\n'
    'These rankings HIDE heterogeneous causal mechanisms across subgroups',
    fontweight='bold')
ax.set_xlabel('Importance')
plt.tight_layout()
plt.savefig('global_importance.png', dpi=150, bbox_inches='tight')
plt.show()
print(f'Top 5 globally: {list(fi.head(5).index)}')
````

---
## 🗺️ Cell 6 — Build the Mapper Lens

We use a **2D lens** combining:
- **Dim 0** → Predicted churn probability (semantic direction: *who is leaving*)
- **Dim 1** → UMAP component 1 or PCA-1 (behavioral structure: *how they differ*)

````python
print('Building lens...')

if HAS_UMAP:
    print('  Running UMAP (n_neighbors=30, min_dist=0.1) — may take ~2 min on 51K rows...')
    reducer = UMAP(n_components=2, n_neighbors=30, min_dist=0.1,
                  random_state=42, n_jobs=-1)
    embedding = reducer.fit_transform(X_scaled)
    structural_dim = embedding[:, 0]
    print(f'  UMAP done. Shape: {embedding.shape}')
else:
    print('  Running PCA...')
    pca = PCA(n_components=2)
    embedding = pca.fit_transform(X_scaled)
    structural_dim = embedding[:, 0]
    print(f'  PCA explained variance: {pca.explained_variance_ratio_[:2].round(3)}')

# Combine into 2D lens and scale to [0,1]
lens_2d = np.column_stack([churn_proba, structural_dim])
lens_2d_scaled = MinMaxScaler().fit_transform(lens_2d)

print(f'\nLens shape: {lens_2d_scaled.shape}')
print('Lens[0]: Churn Probability')
print(f'Lens[1]: {"UMAP-1" if HAS_UMAP else "PC-1"}')
````

---
## 🕸️ Cell 7 — Build the KeplerMapper Graph

````python
# ── Mapper Parameters — properly tuned for Cell2Cell ─────────────────────────
#
# THE PREVIOUS BUG: we were clustering on 52D StandardScaled features with eps=0.3.
# In 52D, average distance between two points ≈ sqrt(2×52) ≈ 10.2.
# eps=0.3 means only points within 3% of avg distance cluster — produces 3-person blobs.
#
# THE FIX: cluster on the 2D UMAP embedding (stored as `embedding` from Cell 6).
# In 2D, distances are interpretable. eps=0.15 on [0,1]-scaled 2D is meaningful.
# This gives us nodes with 50–500 customers each — the right scale for TDA.
#
# Cover parameters:
#   n_cubes=15    → 15×15=225 hypercubes; fine enough for structure, not over-fragmented
#   perc_overlap=0.5 → 50% overlap ensures good connectivity between adjacent cubes
#
# DBSCAN parameters (on 2D UMAP space, scaled to [0,1]):
#   eps=0.15      → ~15% of the embedding range; groups nearby behavioral neighbors
#   min_samples=5 → minimum cluster size; prevents single-point noise nodes

mapper    = km.KeplerMapper(verbose=1)
cover     = Cover(n_cubes=15, perc_overlap=0.5)
clusterer = DBSCAN(eps=0.15, min_samples=5, n_jobs=-1)

# KEY CHANGE: pass the 2D UMAP embedding as the clustering data, not 52D X_scaled
# Both lens and data are now the same 2D space → eps is interpretable
embedding_scaled = MinMaxScaler().fit_transform(embedding)  # scale to [0,1]

graph = mapper.map(
    lens_2d_scaled,          # lens: [churn_proba, umap_1]
    embedding_scaled,        # ← clustering in 2D UMAP space (THE FIX)
    cover=cover,
    clusterer=clusterer
)

n_nodes = len(graph['nodes'])
n_edges = sum(len(v) for v in graph['links'].values()) // 2
print(f'✅ Mapper graph: {n_nodes} nodes | {n_edges} edges')
print(f'   Target: 80–200 nodes, 1–2 connected components')
print(f'   If nodes < 50: decrease eps. If components > 3: increase perc_overlap.')
````

---
## 📐 Cell 8 — Annotate Nodes & Build NetworkX Graph

````python
# Key features to profile per node (raw unscaled values for readability)
PROFILE_FEATS = [
    'monthlyrevenue', 'monthlyminutes', 'customercarecalls',
    'droppedcalls', 'blockedcalls', 'overageminutes',
    'percchangerevenues', 'monthsinservice'
]
PROFILE_FEATS = [f for f in PROFILE_FEATS if f in X_imputed.columns]

# Annotate each node
node_stats = {}
for node_id, members in graph['nodes'].items():
    sub    = df.iloc[members]
    sub_X  = X_imputed.iloc[members]
    stats  = {
        'size':        len(members),
        'churn_rate':  sub[CHURN_COL].mean(),
        'churn_proba': churn_proba[members].mean(),
        'n_churners':  int(sub[CHURN_COL].sum()),
        'members':     members,
    }
    for f in PROFILE_FEATS:
        stats[f'mean_{f}'] = sub_X[f].mean()
    node_stats[node_id] = stats

stats_df = pd.DataFrame(node_stats).T
for col in ['size', 'n_churners', 'churn_rate', 'churn_proba']:
    stats_df[col] = pd.to_numeric(stats_df[col], errors='coerce')

# Build NetworkX
G = nx.Graph()
for node_id, s in node_stats.items():
    G.add_node(node_id, **{k: v for k, v in s.items() if k != 'members'})
for node_id, neighbors in graph['links'].items():
    for nbr in neighbors:
        if not G.has_edge(node_id, nbr):
            sa = set(graph['nodes'][node_id])
            sb = set(graph['nodes'][nbr])
            G.add_edge(node_id, nbr, weight=len(sa & sb) / len(sa | sb))

components = list(nx.connected_components(G))
print(f'NetworkX graph: {G.number_of_nodes()} nodes | {G.number_of_edges()} edges')
print(f'Components: {len(components)} | Largest: {max(len(c) for c in components)} nodes')
print(f'\nChurn rate distribution across nodes:')
print(stats_df['churn_rate'].describe().round(3))
````

---
## 🔥 Cell 9 — Detect Churn Flares

````python
# A churn flare = high churn rate + low graph degree (isolated arm/tip)
# Flare score = churn_rate / (degree + 1)  →  higher = more extreme tip

CHURN_FLARE_THRESHOLD = 0.45   # nodes with >= 45% churn rate
degrees = dict(G.degree())

flare_scores = {}
for node in G.nodes():
    cr = G.nodes[node].get('churn_rate', 0)
    if cr >= CHURN_FLARE_THRESHOLD:
        flare_scores[node] = cr / (degrees[node] + 1)

flare_sorted = sorted(flare_scores.items(), key=lambda x: -x[1])

print(f'Nodes with churn rate >= {CHURN_FLARE_THRESHOLD:.0%}: {len(flare_scores)}')
print()
print(f'{"Node":<30} {"Churn Rate":>12} {"Customers":>10} {"Degree":>8} {"Flare Score":>12}')
print('-' * 75)
for node, score in flare_sorted[:12]:
    nd = G.nodes[node]
    print(f'{node:<30} {nd["churn_rate"]:>12.1%} {int(nd["size"]):>10,} '
          f'{degrees[node]:>8} {score:>12.3f}')
````

---
## 📈 Cell 10 — Interactive Mapper Graph (Plotly)

````python
TOP_N_FLARE = 6
top_flare_ids = [n for n, _ in flare_sorted[:TOP_N_FLARE]]

pos = nx.spring_layout(G, weight='weight', seed=42, k=1.8)

# Edge traces
edge_traces = []
for u, v, d in G.edges(data=True):
    x0, y0 = pos[u]; x1, y1 = pos[v]
    edge_traces.append(go.Scatter(
        x=[x0, x1, None], y=[y0, y1, None], mode='lines',
        line=dict(width=d['weight'] * 2.5,
                  color=f'rgba(180,180,180,{0.2 + d["weight"] * 0.4:.2f})'),
        hoverinfo='none', showlegend=False))

# Node trace
node_x, node_y, node_c, node_s, node_t = [], [], [], [], []
for node in G.nodes():
    nd = G.nodes[node]
    x, y_p = pos[node]
    node_x.append(x); node_y.append(y_p)
    node_c.append(nd.get('churn_rate', 0))
    node_s.append(max(6, min(45, nd.get('size', 1) ** 0.55)))
    lines = [
        f'<b>Node:</b> {node}',
        f'<b>Customers:</b> {int(nd.get("size", 0)):,}',
        f'<b>Churn Rate:</b> {nd.get("churn_rate", 0):.1%}',
        f'<b>Avg Churn Prob:</b> {nd.get("churn_proba", 0):.1%}',
    ]
    for f in PROFILE_FEATS[:4]:
        k = f'mean_{f}'
        if k in nd: lines.append(f'<b>{f}:</b> {nd[k]:.1f}')
    node_t.append('<br>'.join(lines))

node_trace = go.Scatter(
    x=node_x, y=node_y, mode='markers',
    hoverinfo='text', text=node_t,
    marker=dict(
        showscale=True, colorscale='RdYlGn_r', cmin=0, cmax=1,
        color=node_c, size=node_s,
        colorbar=dict(thickness=15, title=dict(text='Churn Rate', side='right'),
                      tickformat='.0%'),
        line=dict(width=1.5, color='white')))

flare_trace = go.Scatter(
    x=[pos[n][0] for n in top_flare_ids if n in pos],
    y=[pos[n][1] for n in top_flare_ids if n in pos],
    mode='markers', name='🔴 Churn Flare',
    marker=dict(symbol='star', size=22, color='red',
                line=dict(color='darkred', width=2)),
    hoverinfo='text',
    text=[f'⭐ FLARE: {n}' for n in top_flare_ids if n in pos])

fig = go.Figure(
    data=edge_traces + [node_trace, flare_trace],
    layout=go.Layout(
        title=dict(
            text='<b>Cell2Cell — KeplerMapper Topological Graph</b><br>'
                 '<sup>Node = customer cluster | Color = Churn Rate | ⭐ = Churn Flare | Hover for profile</sup>',
            x=0.5),
        showlegend=True, hovermode='closest', height=750,
        xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
        yaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
        paper_bgcolor='#0f1117', plot_bgcolor='#0f1117',
        font=dict(color='white', size=12)))
fig.show()
print('🔴 Red arms/flares = concentrated churn  |  🟢 Green core = stable customers')
print('📌 Hover any node to see its feature profile')
````

---
## 🔍 Cell 11 — Flare Feature Profile vs. Global Baseline

````python
# Collect all members across top flare nodes
flare_idx = list(set(
    i for node in top_flare_ids for i in graph['nodes'].get(node, [])
))

flare_means  = X_imputed.iloc[flare_idx][PROFILE_FEATS].mean()
global_means = X_imputed[PROFILE_FEATS].mean()
ratios       = (flare_means / (global_means.replace(0, np.nan) + 1e-9)).sort_values(ascending=False)

churn_rate_flare = df.iloc[flare_idx][CHURN_COL].mean()
print(f'Flare customers:  {len(flare_idx):,}')
print(f'Flare churn rate: {churn_rate_flare:.1%}  (global: {y.mean():.1%})')
print(f'Enrichment:       {churn_rate_flare / y.mean():.1f}x')

# ── Bar chart: deviation from global baseline ─────────────────────────────────
# Color logic:
#   Red   = ratio > 1.1  → feature is ELEVATED in flare vs global (friction signal)
#   Green = ratio < 0.9  → feature is SUPPRESSED in flare vs global (disengagement signal)
#   Gray  = ratio ~1.0   → no meaningful deviation
fig2, ax = plt.subplots(figsize=(10, 5))
ratios_sorted = ratios.sort_values()
bar_colors = [
    '#e74c3c' if r > 1.1 else '#2ecc71' if r < 0.9 else '#95a5a6'
    for r in ratios_sorted.values
]
ratios_sorted.plot(kind='barh', ax=ax, color=bar_colors)
ax.axvline(x=1.0, color='white', linestyle='--', linewidth=1.5, label='= Global average')
ax.set_xlabel('Flare Mean / Global Mean   (ratio)')
ax.set_title(
    'Churn Flare: Feature Deviation from Global Baseline\n'
    '🔴 Red = elevated in flare  |  🟢 Green = suppressed in flare',
    fontweight='bold'
)
ax.legend()

# Annotate with actual values
for j, (feat, ratio) in enumerate(ratios_sorted.items()):
    flare_val  = flare_means.get(feat, 0)
    global_val = global_means.get(feat, 0)
    ax.text(ratio + 0.01, j,
            f'{flare_val:.1f} vs {global_val:.1f}',
            va='center', fontsize=7.5, color='white')

plt.tight_layout()
plt.savefig('flare_profile.png', dpi=150, bbox_inches='tight')
plt.show()

print('\n--- INTERPRETATION ---')
print('Ratios >> 1: feature is HIGHER in flare → friction/escalation signal')
print('Ratios << 1: feature is LOWER in flare  → disengagement/quiet-quit signal')
print()
elevated   = ratios[ratios > 1.1].sort_values(ascending=False)
suppressed = ratios[ratios < 0.9].sort_values()
if not elevated.empty:
    print('Elevated features (potential friction drivers):')
    for feat, r in elevated.items():
        print(f'  {feat:<30} {r:.2f}x global  ({flare_means[feat]:.1f} vs {global_means[feat]:.1f})')
if not suppressed.empty:
    print('Suppressed features (disengagement signals):')
    for feat, r in suppressed.items():
        print(f'  {feat:<30} {r:.2f}x global  ({flare_means[feat]:.1f} vs {global_means[feat]:.1f})')
````

---
## 🎯 Cell 12 — Topological Sample Extraction

````python
flare_df = df.iloc[flare_idx].copy()
flare_df['churn_proba'] = churn_proba[flare_idx]

# Sort by predicted probability — extreme tip = highest probability customers
flare_df = flare_df.sort_values('churn_proba', ascending=False)

TOPO_SAMPLE_SIZE = 300
topo_sample = flare_df.head(TOPO_SAMPLE_SIZE)

print('=' * 55)
print('TOPOLOGICAL SAMPLE — Cell2Cell Churn Flare')
print('=' * 55)
print(f'Total flare members:   {len(flare_df):,}')
print(f'Topological sample:    {len(topo_sample)}')
print(f'Sample churn rate:     {topo_sample[CHURN_COL].mean():.1%}')
print(f'Global churn rate:     {y.mean():.1%}')
print(f'Enrichment:            {topo_sample[CHURN_COL].mean() / y.mean():.1f}x')

print('\nFeature comparison (sample vs global):')
cmp = pd.DataFrame({
    'Flare Sample': topo_sample[PROFILE_FEATS].mean(),
    'Global Mean':  df[PROFILE_FEATS].mean(),
}).round(2)
cmp['Ratio'] = (cmp['Flare Sample'] / cmp['Global Mean']).round(2)
print(cmp.sort_values('Ratio', ascending=False))
````

````python
import json

# Topological sample → Stage 3 LLM causal discovery
topo_sample.to_csv('topo_flare_sample_cell2cell.csv', index=False)

# Full dataset with topology labels
df['churn_proba'] = churn_proba
df['is_in_flare'] = df.index.isin(flare_idx)
df.to_csv('cell2cell_with_topology.csv', index=False)

# Feature descriptions for LLM prompt construction in Stage 3
feature_descriptions = {
    'monthlyrevenue':        'Total monthly revenue billed ($)',
    'monthlyminutes':        'Total call minutes used in the month',
    'customercarecalls':     'Inbound calls to customer care (dissatisfaction proxy)',
    'droppedcalls':          'Number of dropped calls (network quality failure)',
    'blockedcalls':          'Number of blocked calls (network congestion)',
    'overageminutes':        'Minutes used beyond plan limit (pricing friction)',
    'percchangerevenues':    'Month-over-month % change in revenue',
    'percchangeminutes':     'Month-over-month % change in call minutes',
    'monthsinservice':       'Tenure in months',
    'roamingcalls':          'Number of roaming calls',
    'directorassistedcalls': 'Calls requiring directory assistance (friction)',
    'retentioncalls':        'Number of retention calls received',
    'retentionoffersaccepted': 'Number of retention offers accepted',
    'handsetprice':          'Price of current handset ($)',
    'creditrating':          'Credit rating (7=highest, 1=lowest)',
}

metadata = {
    'churn_column': CHURN_COL,
    'feature_cols': FEATURE_COLS,
    'feature_descriptions': feature_descriptions,
    'flare_stats': {
        'n_flare_nodes':          len(top_flare_ids),
        'total_flare_customers':  len(flare_df),
        'topo_sample_size':       TOPO_SAMPLE_SIZE,
        'topo_churn_rate':        float(topo_sample[CHURN_COL].mean()),
        'global_churn_rate':      float(y.mean()),
        'enrichment':             float(topo_sample[CHURN_COL].mean() / y.mean()),
        'top_elevated_features':  ratios.head(5).to_dict(),
    }
}
with open('stage3_metadata.json', 'w') as f:
    json.dump(metadata, f, indent=2)

print('✅ Saved:')
print('  topo_flare_sample_cell2cell.csv  → Stage 3: LLM causal discovery')
print('  cell2cell_with_topology.csv      → Full labelled dataset')
print('  stage3_metadata.json             → Feature descriptions for LLM prompt')
print()
print('=' * 55)
print('PIPELINE:  Stage 1 ✅  Stage 2 ✅  Stage 3 🔜  Stage 4 🔜')
print('=' * 55)
print()
print('➡️  NEXT: Feed stage3_metadata.json into pywhyllm ModelSuggester')
print('   Expected DAG: droppedcalls → customercarecalls → churn')
````

## 🗺️ Pipeline Status

```
✅  Stage 1: Topological Mapping
       └── KeplerMapper on 51K x ~50 real Cell2Cell features
       └── 2D lens: churn_probability + UMAP-1
       └── Churn flares detected and scored

✅  Stage 2: Topological Sampling
       └── 300 customers from flare tip
       └── topo_flare_sample_cell2cell.csv exported

🔜  Stage 3: LLM Causal Discovery  (pywhyllm)
       └── Input: stage3_metadata.json
       └── ModelSuggester → DAG hypothesis
       └── Likely: droppedcalls → customercarecalls → churn
                  overageminutes → churn (separate path)

🔜  Stage 4: Statistical Estimation  (DoWhy)
       └── CausalModel(data=topo_sample, graph=llm_dag)
       └── identify_effect() + estimate_effect()
```
